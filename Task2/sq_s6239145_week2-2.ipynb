```python
# 2.12
from math import atan2, degrees
x = float(input('x: '))
y = float(input('y: '))
angle = atan2(y,x)
print(f'Angle in degrees: {degrees(angle)}')
if x>0 and y>0:
    print('First quadrant')
elif x<0 and y>0:
    print('Second quadrant')
elif x<0 and y<0:
    print('Third quadrant')
elif x>0 and y<0:
    print('Fourth quadrant')
```

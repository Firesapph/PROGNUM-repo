```python
# 3.2
from scipy.constants import G
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

escape_velocities = {}
for i in range(len(names)):
    v_esc = (2 * G * masses[i] / radii[i])**0.5
    print(f'Mass of {names[i]}: {masses[i]:.2f}')
    escape_velocities[names[i]] = v_esc
print(f'Dictionary with escape velocities: {escape_velocities}')
```

```python
# 3.10
from math import pi

# function that determines area and volume given parameters r and h
def cylinder(r, h):
    A = 2*pi*r**2 + 2*pi*r*h
    V = pi*r**2*h
    return A, V
area, volume = cylinder(6, 2)
ratio = area/volume
print(f'ratio area/volume: {ratio}')
```

```python
# 3.17
import numpy as np

# Function that calculates standard deviation from y without using np.std
def stdev(y):
    return (1/len(y) * np.sum(np.array((y-(1/len(y) * np.sum(y)))**2)))**0.5

y = np.array([1, 3, 4, 3, 2])
print(f' Result using my function: {stdev(y)}')
print(f' Result using np.std: {np.std(y)}')
```
